```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
```

```python
df=pd.read_csv("/home/saurav-neupane/full_codes/dynamic-ride-pricing/data/processed/train_features.csv")
```

```python
df.head()
```

```python
X=df.drop(columns=["fare_amount"])
```

```python
y=df["fare_amount"]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, random_state=42, test_size=0.2)
```

```python
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
```

```python
mae=mean_absolute_error(y_test, y_pred)
rmse=root_mean_squared_error(y_test, y_pred)
```

```python
mae
```

```python
rmse
```

MAE measures the average prediction error, while RMSE penalizes large errors more heavily.
The significant gap between MAE and RMSE indicates that the linear model performs reasonably on average but fails on certain edge cases.
This suggests the presence of non-linear relationships and feature interactions that linear regression cannot capture, motivating the use of a Deep Learning model.
